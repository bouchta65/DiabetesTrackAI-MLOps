## 1. Importation des données

```python
import pandas as pd
import matplotlib.pyplot as mpt
import seaborn as sbn

file = r"B:\Machine Learning\Sprint1\YC2_DiabetesTrackAI\data\dataset-diabete-68e2810ab0d7e949117525.csv"
content = pd.read_csv(file)
```

## 2. Aperçu du dataset


```python
#data types
content.info()

#data dimensions
print(content.shape)

#data aperçus
content.head()
```

## 3. Valeurs manquantes et doublons

```python
#valeurs manquantes
print(content.isnull().sum())

#valeur doublons
print(content.duplicated().sum())
```

## 4. Distribution des variables numériques

```python
#Statistiques descriptives
content.describe()
```

```python
# Histogrammes

cols = ["Pregnancies","Glucose","BloodPressure","SkinThickness",
        "Insulin","BMI","DiabetesPedigreeFunction","Age"]



content[cols].hist(figsize=(20,15),color="#1f77b4")
mpt.show()



```

## 5. Corrélations et visualisations


```python
## Pairplot
sbn.pairplot(content[cols],plot_kws={'color':'teal'},diag_kws={'color':'orange'})
mpt.title("relation entre variables")
mpt.show()
```

```python
sbn.heatmap(content[cols].corr(),annot=True, cmap="coolwarm")
mpt.title("Corrélations entre variables")
mpt.show()
```
